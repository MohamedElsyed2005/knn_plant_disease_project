### Visualizing Sample Images

To verify data integrity and understand class appearances, we will display one random sample per class.

```python
import os 
import matplotlib.pyplot as plt 
import glob 
import cv2 
import random

def explore_images(folder_path):
    
      CLASSES = os.listdir(folder_path)
      
      plt.figure(figsize=(15,8))
      for i, cls in enumerate(CLASSES[:15]):
          img_path =  glob.glob(os.path.join(folder_path, cls, "*.jpg"))
          if not img_path:
              print(f"No images found for class {cls}")
              continue
          fb = random.choice(img_path)
          img = cv2.imread(fb)
          img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
          plt.subplot(3, 5, i+1)
          plt.imshow(img_rgb)
          plt.title(cls)
          plt.axis('off')
      plt.suptitle('sample images form each class', fontsize = 18)
      plt.tight_layout(rect=[0,0,1,0.95])
      plt.show()
```

```python
explore_images(r"../data/PlantVillage")
```
